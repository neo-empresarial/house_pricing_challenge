```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
train = pd.read_csv('../data/processed/train.csv')

train['day'] = 1
train['DateSold'] = pd.to_datetime(train.rename(columns={"YrSold": "year",
                                                         "MoSold": "month"})[['year','month','day']],
                                   unit='D')
train['MixArea'] = train[['WoodDeckSF','OpenPorchSF','EnclosedPorch',
                             '3SsnPorch','ScreenPorch','PoolArea']].sum(axis=1)

train.drop(columns=['MiscVal',
                    'GarageType', 'GarageQual',
                    'GarageYrBlt', 'GarageFinish',
                    'GarageCond', 'GarageArea', 'PoolQC',
                    'PoolArea', 'Fireplaces', 'FireplaceQu',
                    'WoodDeckSF', 'day', 'OpenPorchSF',
                    '3SsnPorch', 'ScreenPorch','ExterQual',
                    'ExterCond','BsmtQual', 'KitchenQual'],
           inplace=True)

#train['GarageArea/Car'] = train['GarageArea'] / train['GarageCars']
```
